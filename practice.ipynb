```python
import json
with open("practice.json", "r") as f:
    data = json.load(f)
    print(data)
```

```python
# 2️⃣ Beginner level
print("Company Name:", data["company"])  # Q2
print("First Project Name:", data["projects"][0]["name"])  # Q3
print("First Project Budget:", data["projects"][0]["budget"])  # Q4
print("First Team Member Name:", data["projects"][0]["team"][0]["name"])  # Q5
```

```python
print("All project name and department")
for p in data["projects"]:
    print(p["name"],p["department"])
    
```

```python
print("\nAll Projects - Name & Budget:")  # Q7
for p in data["projects"]:
    print(p["name"], p["budget"])
```

```python

print("\nTeam Sizes per Project:")  # Q8
for p in data["projects"]:
    print(p["name"], len(p["team"]))
```

```python
# Q9: Projects with budget > 100000
print("\nProjects with budget > 100000:")
for p in data["projects"]:
    if p["budget"] > 100000:
        print(p["name"], p["budget"])
```

```python
# Q10: Simple budget level column
for p in data["projects"]:
    if p["budget"] > 100000 :
        p["budget_level"] = "high"
    else:
        p["budget_level"] = "low"
```

```python
print(data["projects"][0]["budget_level"])  # Verify budget_level addition
```

```python
# 3️⃣ Medium Level
import pandas as pd
projects_df = pd.json_normalize(data["projects"], sep="_")  # Q11
```

```python
print(projects_df)
```

```python
team_df = pd.json_normalize(
    data["projects"],
    record_path="team",
    meta=["project_id", "name", "department", "budget", "budget_level"],
    record_prefix="team_",   # prefix for fields from team
    meta_prefix="project_",  # prefix for fields from project
)
```

```python
print(team_df)
```

```python
# Q13: Total team members per department
team_count_dept = team_df.groupby("project_department")["team_id"].count().reset_index(name="total_team_members")
print("\nTeam Members per Department:\n", team_count_dept)
```

```python
team_count_dept = team_df.groupby("project_department")["team_id"].count().reset_index(name="total_team_members")
print("\nTeam Members per Department:\n", team_count_dept)
```
