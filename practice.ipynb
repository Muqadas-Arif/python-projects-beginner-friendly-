```python
import json
with open("practice.json", "r") as f:
    data = json.load(f)
    print(data)
```

```python
print(data["company"])
```

```python
print(data["employees"][2]["address"]['city'])
```

```python
for emp in data["employees"]:
    print(emp["name"],emp["department"],emp["salary"])
```

```python
import pandas as pd
df=pd.json_normalize(data["employees"],sep='_')
print(df)
```

How many employees are in each department?

```python
df.groupby("department")["id"].count().reset_index(name="employee_count")
```

Average salary per department



```python
df.groupby("department")["salary"].mean().reset_index(name="average_salary")
```

Salary distribution (CASE WHEN logic)

```python
df["salary"].value_counts()
```

```python
df["eligible"] = (
    ((df["department"] == "IT") & (df["salary"] >= 70000)) |
    ((df["department"] == "HR") & (df["salary"] >= 55000))
)
```

```python
print(df["eligible"])
```
